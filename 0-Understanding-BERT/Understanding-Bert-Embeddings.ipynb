```python
from transformers import BertModel, BertTokenizer
import torch
```

```python
sentence = "I like to workout everyday."
```

## **Extracting Embeddings from Pre-Trained BERT**

### **Tokenization**

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
tokens = tokenizer.tokenize(sentence)
print(tokens)
```

Add the [CLS] token (classifer) at the beginning and [SEP] token (separator) at the end of the tokens list.

```python
tokens = ['[CLS]'] + tokens + ['[SEP]']
print(tokens)
```

```python
print(len(tokens))
print(tokens)
```

```python
# ADD PAD TOKENS

tokens = tokens + ['[PAD]'] + ['[PAD]']
print(tokens)
```

```python
attention_mask = [1 if i!= '[PAD]' else 0 for i in tokens]
print(attention_mask)
```

### **Token IDs**

```python
token_ids = tokenizer.convert_tokens_to_ids(tokens)
print(token_ids)
```

```python
token_ids = torch.tensor(token_ids).unsqueeze(0)

attention_mask = torch.tensor(attention_mask).unsqueeze(0)
```

### **Embeddings**

Embeddings from the top-most encoder layer (12th encoder)

```python
model = BertModel.from_pretrained('bert-base-uncased')
```

```python
output = model(
    token_ids,
    attention_mask = attention_mask
)

output
```
